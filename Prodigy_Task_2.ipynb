```python
# ==========================
# EDA Visualizations (Task-02)
# ==========================

import matplotlib.pyplot as plt
import pandas as pd

# 1. Gender Distribution
gender = df["Sex"].value_counts()

plt.figure(figsize=(6,5))
plt.bar(gender.index, gender.values, color=["hotpink", "deepskyblue"], edgecolor="black")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# 2. Age Distribution
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, color="pink", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# 3. Passenger Class Distribution
pclass = df["Pclass"].value_counts().sort_index()

plt.figure(figsize=(6,5))
plt.bar(pclass.index.astype(str), pclass.values,
        color=["#FF9999","#FF66B2","#CC66FF"],
        edgecolor="black")
plt.title("Passenger Class Distribution")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# 4. Survival by Gender
survival_gender = pd.crosstab(df["Sex"], df["Survived"])

survival_gender.plot(kind="bar",
                     figsize=(7,5),
                     color=["lightcoral","mediumseagreen"],
                     edgecolor="black")

plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# 5. Survival by Passenger Class
survival_class = pd.crosstab(df["Pclass"], df["Survived"])

survival_class.plot(kind="bar",
                    figsize=(7,5),
                    color=["tomato","limegreen"],
                    edgecolor="black")

plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# 6. Fare Distribution
plt.figure(figsize=(8,5))
plt.hist(df["Fare"], bins=30, color="violet", edgecolor="black")
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# 7. Box Plot - Age
plt.figure(figsize=(6,4))
plt.boxplot(df["Age"], vert=False, patch_artist=True,
            boxprops=dict(facecolor="pink"))
plt.title("Box Plot of Age")
plt.xlabel("Age")
plt.show()

# 8. Box Plot - Fare
plt.figure(figsize=(6,4))
plt.boxplot(df["Fare"], vert=False, patch_artist=True,
            boxprops=dict(facecolor="plum"))
plt.title("Box Plot of Fare")
plt.xlabel("Fare")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create 3x3 dashboard
fig, axes = plt.subplots(3, 3, figsize=(18, 15))
fig.suptitle("Train Dataset - Exploratory Data Analysis Dashboard",
             fontsize=20, fontweight='bold')

# 1. Gender Distribution
gender = df["Sex"].value_counts()
axes[0,0].bar(gender.index, gender.values,
              color=["hotpink", "deepskyblue"], edgecolor="black")
axes[0,0].set_title("Gender Distribution")
axes[0,0].set_xlabel("Sex")
axes[0,0].set_ylabel("Count")

# 2. Age Distribution
axes[0,1].hist(df["Age"].dropna(),
               bins=20,
               color="pink",
               edgecolor="black")
axes[0,1].set_title("Age Distribution")
axes[0,1].set_xlabel("Age")
axes[0,1].set_ylabel("Frequency")

# 3. Passenger Class Distribution
pclass = df["Pclass"].value_counts().sort_index()
axes[0,2].bar(pclass.index.astype(str),
              pclass.values,
              color=["#FF9999", "#FF66B2", "#CC66FF"],
              edgecolor="black")
axes[0,2].set_title("Passenger Class Distribution")
axes[0,2].set_xlabel("Class")
axes[0,2].set_ylabel("Count")

# 4. Survival Distribution
survival = df["Survived"].value_counts().sort_index()
axes[1,0].bar(["Not Survived","Survived"],
              survival.values,
              color=["tomato","limegreen"],
              edgecolor="black")
axes[1,0].set_title("Survival Distribution")
axes[1,0].set_ylabel("Count")

# 5. Survival by Gender
survival_gender = pd.crosstab(df["Sex"], df["Survived"])
survival_gender.plot(kind="bar",
                     ax=axes[1,1],
                     color=["lightcoral","mediumseagreen"],
                     edgecolor="black")
axes[1,1].set_title("Survival by Gender")
axes[1,1].set_xlabel("Sex")
axes[1,1].set_ylabel("Count")
axes[1,1].tick_params(axis='x', rotation=0)

# 6. Survival by Passenger Class
survival_class = pd.crosstab(df["Pclass"], df["Survived"])
survival_class.plot(kind="bar",
                    ax=axes[1,2],
                    color=["orange","green"],
                    edgecolor="black")
axes[1,2].set_title("Survival by Passenger Class")
axes[1,2].set_xlabel("Class")
axes[1,2].set_ylabel("Count")
axes[1,2].tick_params(axis='x', rotation=0)

# 7. Fare Distribution
axes[2,0].hist(df["Fare"],
               bins=30,
               color="violet",
               edgecolor="black")
axes[2,0].set_title("Fare Distribution")
axes[2,0].set_xlabel("Fare")
axes[2,0].set_ylabel("Frequency")

# 8. Age Box Plot
axes[2,1].boxplot(df["Age"].dropna(),
                  vert=False,
                  patch_artist=True,
                  boxprops=dict(facecolor="pink"))
axes[2,1].set_title("Age Box Plot")
axes[2,1].set_xlabel("Age")

# 9. Fare Box Plot
axes[2,2].boxplot(df["Fare"],
                  vert=False,
                  patch_artist=True,
                  boxprops=dict(facecolor="plum"))
axes[2,2].set_title("Fare Box Plot")
axes[2,2].set_xlabel("Fare")

# Improve layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Save dashboard (optional)
plt.savefig("Titanic_EDA_Dashboard.png", dpi=300)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dashboard
fig, axes = plt.subplots(3, 3, figsize=(18, 15))
fig.suptitle("Train Dataset - Exploratory Data Analysis Dashboard",
             fontsize=20, fontweight='bold')

# 1. Gender Distribution
gender = df["Sex"].value_counts()
axes[0,0].bar(gender.index, gender.values,
              color="black", edgecolor="black")
axes[0,0].set_title("Gender Distribution")
axes[0,0].set_xlabel("Sex")
axes[0,0].set_ylabel("Count")

# 2. Age Distribution
axes[0,1].hist(df["Age"].dropna(),
               bins=20,
               color="gray",
               edgecolor="black")
axes[0,1].set_title("Age Distribution")
axes[0,1].set_xlabel("Age")
axes[0,1].set_ylabel("Frequency")

# 3. Passenger Class Distribution
pclass = df["Pclass"].value_counts().sort_index()
axes[0,2].bar(pclass.index.astype(str),
              pclass.values,
              color="dimgray",
              edgecolor="black")
axes[0,2].set_title("Passenger Class Distribution")
axes[0,2].set_xlabel("Class")
axes[0,2].set_ylabel("Count")

# 4. Survival Distribution
survival = df["Survived"].value_counts().sort_index()
axes[1,0].bar(["Not Survived", "Survived"],
              survival.values,
              color="gray",
              edgecolor="black")
axes[1,0].set_title("Survival Distribution")
axes[1,0].set_ylabel("Count")

# 5. Survival by Gender
survival_gender = pd.crosstab(df["Sex"], df["Survived"])
survival_gender.plot(kind="bar",
                     ax=axes[1,1],
                     color=["black", "lightgray"],
                     edgecolor="black")
axes[1,1].set_title("Survival by Gender")
axes[1,1].set_xlabel("Sex")
axes[1,1].set_ylabel("Count")
axes[1,1].tick_params(axis='x', rotation=0)
axes[1,1].legend(["Not Survived", "Survived"])

# 6. Survival by Passenger Class
survival_class = pd.crosstab(df["Pclass"], df["Survived"])
survival_class.plot(kind="bar",
                    ax=axes[1,2],
                    color=["black", "lightgray"],
                    edgecolor="black")
axes[1,2].set_title("Survival by Passenger Class")
axes[1,2].set_xlabel("Class")
axes[1,2].set_ylabel("Count")
axes[1,2].tick_params(axis='x', rotation=0)
axes[1,2].legend(["Not Survived", "Survived"])

# 7. Fare Distribution
axes[2,0].hist(df["Fare"],
               bins=30,
               color="gray",
               edgecolor="black")
axes[2,0].set_title("Fare Distribution")
axes[2,0].set_xlabel("Fare")
axes[2,0].set_ylabel("Frequency")

# 8. Age Box Plot
axes[2,1].boxplot(df["Age"].dropna(),
                  vert=False,
                  patch_artist=True,
                  boxprops=dict(facecolor="lightgray", color="black"),
                  medianprops=dict(color="black"),
                  whiskerprops=dict(color="black"),
                  capprops=dict(color="black"))
axes[2,1].set_title("Age Box Plot")
axes[2,1].set_xlabel("Age")

# 9. Fare Box Plot
axes[2,2].boxplot(df["Fare"],
                  vert=False,
                  patch_artist=True,
                  boxprops=dict(facecolor="lightgray", color="black"),
                  medianprops=dict(color="black"),
                  whiskerprops=dict(color="black"),
                  capprops=dict(color="black"))
axes[2,2].set_title("Fare Box Plot")
axes[2,2].set_xlabel("Fare")

# Layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Save the dashboard (optional)
plt.savefig("train_EDA_Dashboard_BW.png", dpi=300)

plt.show()
```
